```python
import json, re
```

```python
def brushing_json_str(x: str)  -> str:
    # convert enclosing strings with double quotes and
    # treat the case if JSON holds escaped single-quotes (\')
    escaped_s_quotes = re.compile('(?<!\\\\)\'')
    none_sub = re.compile('None')
    inner_db_quotes = re.compile('(?<=[a-zA-z0-9])(\s*\"\s*)(?=[a-zA-z0-9])')
    inner_quoted_json_start = re.compile('":"{') 
    inner_quoted_json_end = re.compile('}","')
    quoted_json_end = re.compile('}"')
    json_str = escaped_s_quotes.sub('\"', x)
    json_str = inner_db_quotes.sub("\'", json_str)
    json_str = none_sub.sub('null', json_str)
    json_str = inner_quoted_json_start.sub('":{', json_str)
    json_str = inner_quoted_json_end.sub('},"', json_str)
    json_str = quoted_json_end.sub('}', json_str)
    return json_str
```

```python
with open("query_result_2022-04-26T08_39_39.505263Z.json") as f:
    js = json.load(f)
```

```python
res = brushing_json_str(js)
```
